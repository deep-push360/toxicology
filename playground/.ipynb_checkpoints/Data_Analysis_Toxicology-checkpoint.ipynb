```python
## prepare data for model

def data_generator(dataset):
    """
   a function to return x and y
   where
   x = list of comments
   y = a one hot vector of the classes
   """

    # remove punctuation
    dataset['comment_text'] = [re.sub('[^\w\s\']|(\n)',' ', i) for i in dataset['comment_text']]
    dataset['comment_text'] = [re.sub( '\s+', ' ', i).strip() for i in dataset['comment_text']]

    # create list and array
    x = list(dataset['comment_text'])
    y = np.array(dataset.iloc[:,2:])
    
    return[x,y]
```

```python
# Data Exploratory

# add feature: comment length (number of character - removing spaces)

total_df['length'] = [len(str(i).replace(' ','')) for i in total_df['comment_text']]

# add feature: count capitallize letters

total_df['cap_count'] = [len(re.findall(r'[A-Z]',str(i))) for i in total_df['comment_text']]

# add feature: count number of question/exclamation marks

total_df['marks_count'] = [len(re.findall(r'\?|\!',str(i))) for i in total_df['comment_text']]
```
